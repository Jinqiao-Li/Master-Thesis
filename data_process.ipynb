## Defined functions

I defined two functions here for dataframe computation.

```python

def get_GWA_IWA(dwa_id):
    # This function is used to get the corresponding GWA ID(Element ID), GWA title(Element Name), IWA ID and IWA Title
    # param: DWA ID
    gwa_iwa = dwa.loc[dwa['DWA ID']==dwa_id][['Element ID','Element Name','IWA ID', 'IWA Title']].values

    return gwa_iwa.flatten()


def get_de(task_id, gwa_id, iwa_id, dwa_id):
    # This function is defined 
    task_de = onet_de.loc[onet_de['id']==str(task_id)]['text'].values
    gwa_de = onet_de.loc[onet_de['id']==str(gwa_id)]['text'].values
    iwa_de = onet_de.loc[onet_de['id']==str(iwa_id)]['text'].values
    dwa_de = onet_de.loc[onet_de['id']==str(dwa_id)]['text'].values
    return task_de,gwa_de,iwa_de,dwa_de
```

```python

# get the corresponsing GWA and IWA info (id and text)
task_to_dwa[['GWA ID','GWA Title','IWA ID','IWA Title']] = task_to_dwa.apply(lambda x: pd.Series(get_GWA_IWA(x['DWA ID'])),axis=1)

# get the German version of task, gwa, iwa, dwa from onet by id
task_to_dwa[['Task_de','GWA_de','IWA_de','DWA_de']] = task_to_dwa.apply(
    lambda x: pd.Series(get_de(x['Task ID'],x['GWA ID'], x['IWA ID'], x['DWA ID'])),axis=1)
```
